```python
def getRanking(query, dataframe_empty):
    google_url = 'https://www.google.com/search'
    my_params = {'q': query}
    my_headers = getheaders()
    print(my_headers)
    r = requests.get(google_url, params = my_params, headers = my_headers)
    now = int(round(time.time()*1000))
    now_format = time.strftime('%Y-%m-%d %H:%M:%S',time.localtime(now/1000))
    if r.status_code == requests.codes.ok:
      # Analysis HTML Original Code by BeautifulSoup
        soup = BeautifulSoup(r.text, 'html.parser')

      # Oberserve HTML Original Code
        #print(soup.prettify())

      # Scrape search results by CSS selector
        items = soup.select('.r > a[href]')
        #  items = soup.select('.r > a[href^="/url"]')   
       #rso > div:nth-child(1) > div > div > div > div.r > a
        #items = soup.select('rso > div:nth-of-type(4) > div > div:nth-of-type(1) > div > div > div.r > a')
        print(len(items))
    for i in items:
    # Title
        print("Title：" + i.text)
    # Url
        print("Url：" + i.get('href'))
        
    
    ranking = 1
    for i in items:
        Url = i.get('href')
        dataframe_empty= dataframe_empty.append(pd.DataFrame({'Keyword':[query],'Ranking':[ranking], 'URL': [Url], 'Date' : [now_format]}), ignore_index=True)
        ranking = ranking+1
        
    return dataframe_empty

```

```python
for k in keywords_list:
    dataframe_empty = getRanking(k, dataframe_empty)
    time.sleep(random.randrange(3,60))
export_csv = dataframe_empty.to_csv (r'/Users/hipposon/Postgraduate_Project/File/export_dataframe.csv', index = None, header=True)#Don't forget to add '.csv' at the end of the path
print("done")
```
